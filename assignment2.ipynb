```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train = pd.read_csv('/content/train_loan (1).csv')
```

```python
df = train.copy()
```

```python
df.head()
```

```python
#for finding missing values
df.isna().sum()
```

```python
# Replace NaNs with mode for categorical columns
df['Gender'].fillna(df['Gender'].mode()[0], inplace=True)
df['Married'].fillna(df['Married'].mode()[0], inplace=True)
df['Dependents'].fillna(df['Dependents'].mode()[0], inplace=True)
df['Self_Employed'].fillna(df['Self_Employed'].mode()[0], inplace=True)
df['Loan_Amount_Term'].fillna(df['Loan_Amount_Term'].mode()[0], inplace=True)
```

```python
# Impute missing values in numerical columns with median
df['LoanAmount'].fillna(df['LoanAmount'].median(), inplace=True)
```

```python
# Impute Credit_History with a new category
df['Credit_History'].fillna('No Credit History', inplace=True)
```

```python
df.isna().sum()
```

```python
df.info()
```

```python
# Perform one-hot encoding for categorical variables
df = pd.get_dummies(df, columns=['Gender', 'Married', 'Dependents', 'Education', 'Self_Employed', 'Property_Area'])
```

```python
# Encode Loan_Status separately since it's the target variable
df['Loan_Status'] = df['Loan_Status'].map({'Y': 1, 'N': 0})
```

```python
# Encode Credit_History using label encoding since it's binary
df['Credit_History'] = df['Credit_History'].map({'Yes': 1, 'No': 0, 'No Credit History': -1})
```

```python
df.head()
```

```python
# Drop rows with NaN values in the 'Credit_History' column
df.dropna(subset=['Credit_History'], inplace=True)
```

```python
# Reset the index after dropping rows
df.reset_index(drop=True, inplace=True)
```

```python
df.head()
```

```python
#drop loan_id from dataset
df.drop(columns=['Loan_ID'], inplace=True)
```

```python
# Split data into features and target variable
x = df.drop(columns=['Loan_Status'])
y = df['Loan_Status']
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Split data into training and testing sets
x_train,x_test,y_train,y_test = train_test_split(x.values,y,test_size = 0.3,random_state = 42)
```

```python
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from catboost import CatBoostRegressor
```

```python
#initialize the models
models = {'Random Forest': RandomForestRegressor(), 'XGB Regressor': XGBRegressor(),'Catboost': CatBoostRegressor(verbose=0)}
```

```python
#train the models
for name,model in models.items():
  model.fit(x_train,y_train)
  print(name + 'trained')
```

```python
from sklearn.metrics import mean_squared_error
```

```python
#finding the mse
for name,model in models.items():
  y_pred = model.predict(x_test)
  mse = mean_squared_error(y_test,y_pred)
  print(name + 'MSE {:.2f}'.format(mse))
```
